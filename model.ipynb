```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow.keras as keras
```

```python
batch = 64
train_output = np.ones((64,400))
train_input = np.zeros((64,400))
```

```python
for j in range(batch):
    for i in range(0,400,20):
        train_input[j,i]=1
```

```python
model = keras.Sequential((
    keras.Input(shape=(400,)),
    keras.layers.Dense(400,activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(800,activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1200,activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1200,activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(800,activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(400,activation="sigmoid")    
))
```

```python
model.compile(
    optimizer="Adam",
    loss=keras.losses.categorical_crossentropy,
    metrics="accuracy"
)
```

```python
model.summary()
```

```python
model.fit(train_input,train_output,epochs=50000)
```
